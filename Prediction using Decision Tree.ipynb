## Task 6 : Prediction using Decision Tree Algorithm

### Create the Decision Tree classifier and visualize it graphically. 

### Step 1 : Importing libraries

```python
import sklearn.datasets as datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import tree
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2 : Importing data from the source

```python
iris=datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

### Step 3 : Displaying the data

```python
X.head()
```

```python
X.tail()
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
Y = iris.target
```

### Step 4 : Data Visualization to compare data

```python
sns.pairplot(X)
```

### Step 5 : Train data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

### Step 6 :  Decision tree model training

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_predict = dtc.predict(X_test)
```

### Step 7 : Visualizaing the trained model

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (6,4), dpi = 700)
tree.plot_tree(dtc, feature_names = fn, class_names = cn, filled = True);
```
